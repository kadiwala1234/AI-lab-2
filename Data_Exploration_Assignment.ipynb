```python
# Data Manipulation with Pandas

import pandas as pd

# Creating a sample dataset
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 22],
    "Score": [85, 90, 78]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display DataFrame
df
```

## Dataset Description

The dataset contains basic information about three individuals:
- Name: Person's name
- Age: Person's age
- Score: Their performance score

This dataset will be used to demonstrate data manipulation and visualization.

```python
# Data Visualization

import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(6,4))
sns.barplot(x="Name", y="Score", data=df)

plt.title("Scores of Individuals")
plt.xlabel("Name")
plt.ylabel("Score")

plt.show()
```

## Data Visualization Explanation

The bar chart shows the scores of each individual in the dataset.
- Bob has the highest score.
- Charlie has the lowest score.

This visualization helps quickly compare performance across individuals.
